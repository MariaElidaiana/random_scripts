```python
import arxivscraper.arxivscraper as ax
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Scrapping Arxiv for XENON1T on the abstract 

scraper = ax.Scraper(category='physics:hep-ph',date_from='2020-06-17',date_until='2020-06-27',t=10, filters={'abstract':['XENON1T']})
output = scraper.scrape()
```

```python
#Checking the papers

cols = ['id', 'title', 'categories', 'abstract', 'doi', 'created', 'updated', 'authors']
df = pd.DataFrame(output,columns=cols)

df 
```

```python
#Remove the first two paper that are not related

df = df.iloc[2:]

df
```

```python
plt.figure(figsize=(8,8))
df.groupby('created')['id'].nunique().plot(kind='bar')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Number of XENON1T papers on Arvix', fontsize=20)
plt.xticks(rotation=45, fontsize=20)
plt.yticks(fontsize=20)
plt.show()
```
